```python
# 导入 AutoTokenizer 和 AutoModel 类
from transformers import AutoTokenizer, AutoModel

# 使用 AutoModel 类的 from_pretrained 方法加载一个预训练的 BERT 模型
model = AutoModel.from_pretrained("bert-base-uncased")

# 打印加载的模型
print(model)
```

```python
print(model.num_parameters())
```

```python
# 从 transformers 库中导入 AutoTokenizer 类
from transformers import AutoTokenizer

# 使用 AutoTokenizer 类的 from_pretrained 方法加载一个预训练的 BERT 分词器
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# 待处理的原始文本列表
raw_inputs = [
    "I've been waiting for a HuggingFace course my whole life.",
    "I hate this so much!",
]

# 使用加载的分词器对原始文本进行处理，包括分词、填充、截断等操作，并返回结果作为 PyTorch 张量
inputs = tokenizer(raw_inputs, padding=True, truncation=True, return_tensors="pt")

# 打印处理后的结果
print(inputs)
```
